### `Question 2 - Find the top 5 superpowers in descending order

```python
def string_to_list(string):
    # Remove square brackets and quotes, split by comma, and strip whitespace
    elements= string.strip('[]').replace('\'', '').split(',')
    return [elem.strip() for elem in elements]

df['superpowers'] = df['superpowers'].apply(lambda x: list(string_to_list(x)))
from collections import Counter


# Flatten the arrays into a single list
flat_list = [item for sublist in df['superpowers'] for item in sublist]

# Count the distinct superpower in the flattened list
distinct_superpower_counts = dict(Counter(flat_list))

sorted_items = sorted(distinct_superpower_counts.items(), key=lambda item: item[1], reverse=True)
n_largest_items = sorted_items[:5]

# Display the '5' largest items with their counts
for item, count in n_largest_items:
    print(f"{item}: {count}")
```

### `Question 8 - Report on the 10 superheros with most relative

```python
import re
df_relatives = df[['name','alignment','relatives']]
df_relatives['relatives']=df_relatives['relatives'].fillna(value='')
df_relatives['alignment']=df_relatives['alignment'].fillna(value='None')

def count_distinct_relatives(group): 
    
    #elements= group.strip('[]').replace('\'', '').replace('\n', ';').replace(')', ';').split(';') 
    
    group = re.sub(r'\(.+?\)',';',group)   
    elements= group.strip('.').replace(',', '')
    elements= re.split(';',elements)  
    distinct_teams = pd.Series(elements).unique()
    return (len([elem.strip() for elem in distinct_teams if elem.strip() != '']))



df_relatives['numberofrelatives'] = df_relatives['relatives'].apply(count_distinct_relatives)
df_relatives.sort_values(by='numberofrelatives', ascending=False).head(10)
```
